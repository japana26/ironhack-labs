### Police Data Set

```python
original_police_db = pd.read_csv('Policias.csv')
original_police_db.head(3)
```

```python
#selecting the only columns that we need
police_db = original_police_db[['NOME', 'MORADA','CODPOSTAL']]
police_db.head(5)
```

```python
#checking with lines have no postcode
nopostcode = police_db[police_db['CODPOSTAL'] == '-']
nopostcode
```

```python
police_db.shape
```

```python
#find duplicated addresses and remove them. Consider these as one police division
police_db.drop_duplicates(subset ="MORADA", keep = False, inplace = True) 
police_db
```

```python
#drop all rows with - on postal code
police_db.drop(police_db[police_db['CODPOSTAL'] == '-'].index, axis=0, inplace=True)
```

```python
#resetting the index
police_db.reset_index(drop=True, inplace=True)
police_db
```

```python
#creating new columns for Freguesias
police_db['Freguesia'] = police_db['CODPOSTAL']
```

```python
#dropping police estalishments that are not stations
police_db.drop(police_db[police_db['NOME'] == 'Secção de Achados'].index, axis=0, inplace=True)
```

```python
#dropping police estalishments that are not stations
police_db.drop(police_db[police_db['NOME'] == 'Instituto Superior de Ciências Policiais e Segurança Interna'].index, axis=0, inplace=True)
police_db.drop(police_db[police_db['NOME'] == 'Parque de Viaturas Rebocadas'].index, axis=0, inplace=True)
police_db.drop(police_db[police_db['NOME'] == 'Carreira de Tiro da Escola Superior de Polícia'].index, axis=0, inplace=True)
```
