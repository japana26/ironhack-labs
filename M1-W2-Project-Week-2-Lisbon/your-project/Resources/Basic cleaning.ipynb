# Basic cleaning

```python
import pandas as pd
```

## Cleaning Population DS

```python
original_pop_db = pd.read_excel('population.xlsx')
original_pop_db.head(10)
```

```python
# Selecting just the cols that we want to display

pop_db = original_pop_db[['GEO_COD_DSG', 'NIVEL_DSG', 'N_INDIVIDUOS_RESIDENT']]

pop_db = pop_db[(pop_db['NIVEL_DSG'] == 'Freguesia')]

pop_db = pop_db[['GEO_COD_DSG', 'N_INDIVIDUOS_RESIDENT']]

pop_db.head(10)
```

```python
# Renaming the columns

pop_db.columns = ['FREGUESIAS', 'TOTAL RESIDENTS']

pop_db['FREGUESIAS'] = pop_db['FREGUESIAS'].str.upper()

pop_db.head(10)
```

```python
pop_db.shape
```

## Cleaning Freguesias DS

```python
# Importing the file

original_freguesias_db = pd.read_csv('Freguesias2012.csv')
original_freguesias_db.head(20)
```

```python
# Selecting the cols that we want to keep

freguesia_db = original_freguesias_db[['NOME', 'AREA_M2', 'FREGUESIAS53']]
freguesia_db.head(10)
```

```python
# Renaming the cols

freguesia_db.columns = ['CURRENT FREGUESIAS', 'AREA', 'OLD FREGUESIAS']

freguesia_db['CURRENT FREGUESIAS'] = freguesia_db['CURRENT FREGUESIAS'].str.upper()

freguesia_db.head(10)
```

## Cleaning Public Lighting DS
